```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/diabetes.csv',index_col='Pregnancies')
data
```

```python
data.head()
```

```python
def myScatter (label, color) :
  la_data = data.loc[label]
  plt.scatter(la_data['Glucose'],
              la_data['BloodPressure'],
              c = color,
              label = label)
```

```python
plt.figure(figsize=(5,5))
myScatter(4,'red')
myScatter(1,'blue')
myScatter(2,'yellow')
myScatter(3,'green')

plt.legend()
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.show
```

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
data[['Glucose','BloodPressure']]
```

```python
x = data.iloc[:,2:]
y = data.iloc[:,0]
```

```python
x_train = x.iloc[:350,:]
x_test = x.iloc[350:,:]
y_train = y[:350]
y_test = y[350:]
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=3)
```

```python
knn_model.fit(x_train,y_train)
```

```python
pre = knn_model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(pre, y_test)
```
